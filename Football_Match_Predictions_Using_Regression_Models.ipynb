# Predicting Football Match Outcomes Using Regression Models 

## 1. Defining the Question

### a) Specifying the Question



1.   Predict the number of goals scored by the home team
2.   Predict the number of goals scored by the away team
3.   Predict the outcome of a football match between team 1 and team 2





### b) Defining the Metric for Success



1.   A polynomial regression model that predicts the number of goals scored by the home team with above 50% accuracy
2.   A polynomial regression model that predicts the number of goals scored by the away team with an accuracy of above 50%
3.   A logistic regression model that predicts the outcome of a football match between team 1 and team 2
4.   Quantile, lasso and ridge regression models that predict the outcome of a football match








### c) Understanding the context 

The venue and ranking of national teams in football plays a role in the outcome of football matches. 

Teams that play at home are more likely to win than away teams.

Home teams are also more likely to score more goals than away teams.

```
round_pre = [np.round(value) for value in polpre]

print('Quad RMSE: ', np.sqrt(mean_squared_error(lebo_train, round_pre)), '\nQuad r_squared: ', quad_r2)
```

```
# getting the residuals to check for heteroskedasticity 
resid = np.subtract(polpre, lebo_train)
pd.DataFrame(resid)

resid.mean()
```

```
# plotting the residuals 

plt.scatter(polpre, resid, color='black')
plt.ylabel('residual')
plt.xlabel('fitted values')
plt.axhline(y= resid.mean(), color='red', linewidth=1)
plt.show()
```

```
import scipy as sp

test_result, p_val = sp.stats.bartlett(polpre.flatten(), resid.flatten())

degree_of_f = len(polpre)-1
prob = 1 - p_val

critical_value = sp.stats.chi2.ppf(prob, degree_of_f)
print(p_val)

# If the test_result is greater than the critical value, then we reject our null
# hypothesis. This would mean that there are patterns to the variance of the data

# Otherwise, we can identify no patterns, and we accept the null hypothesis that 
# the variance is homogeneous across our data

if (test_result > critical_value):
  print('the variances are unequal, and the model should be reassessed')
else:
  print('the variances are homogeneous!')

```

```
lincub = lin.fit(x_cube, lebo_train)

# making predictions
y_cube_fit = lincub.predict(cubic.fit_transform(feat_fit))

# calculating R squared
# quad_r2 = r2_score(lebo_test, linpol.predict(x_quad))
cubpre = linpol.predict(x_cube)
cub_r2 = r2_score(lebo_train, cubpre)

print('Cubic RMSE: ', np.sqrt(mean_squared_error(lebo_train, cubpre)), '\nCubic r_squared: ', cub_r2)
```

```
# plotting the residuals 

plt.scatter(cubpre, resid, color='black')
plt.ylabel('residual')
plt.xlabel('fitted values')
plt.axhline(y= resid.mean(), color='red', linewidth=1)
plt.show()
```

```
# determining heteroskedasticity using bartlet's test

test_res, p_valu = sp.stats.bartlett(cubpre.flatten(), resid.flatten())

degree_of_fre = len(cubpre)-1
proba = 1 - p_valu

critical_valu = sp.stats.chi2.ppf(proba, degree_of_fre)
print(p_valu)

# If the test_result is greater than the critical value, then we reject our null
# hypothesis. This would mean that there are patterns to the variance of the data

# Otherwise, we can identify no patterns, and we accept the null hypothesis that 
# the variance is homogeneous across our data

if (test_res > critical_valu):
  print('the variances are unequal, and the model should be reassessed')
else:
  print('the variances are homogeneous!')

```

```
# determining heteroskedasticity using bartlet's test

tes_res, p_val = sp.stats.bartlett(apolpre.flatten(), resds.flatten())

deg_of_f = len(apolpre)-1
pro = 1 - p_val

crit_val = sp.stats.chi2.ppf(pro, deg_of_f)
print(p_val)


if (tes_res > crit_val):
  print('the variances are unequal, and the model should be reassessed')
else:
  print('the variances are homogeneous!')

  
# If the test_result is greater than the critical value, then we reject our null
# hypothesis. This would mean that there are patterns to the variance of the data

# Otherwise, we can identify no patterns, and we accept the null hypothesis that 
# the variance is homogeneous across our data
```

### Quantile Regression

```
# quantile regression works well with data having outliers, heteroskedasticity and skewness
# fin_com dataframe has outliers and has skewness in the home and away scores
fin_com.head()
```

```
# visualizing a pairplot to help in picking the variables suitable for quantile regression

sns.pairplot(fin_com, dropna=True)

# pairplot reveals home score, away score and home and away rank have skewness 
# these variables can be used in a quantile regression model
```

```
# using home rank and hone_score columns as they are both skewed to the right
mood = smf.quantreg('home_score ~ home_rank', fin_com)
rslt = mood.fit(q=.5)
print(rslt.summary())
```

home score column has a very small regression coefficient

```
percentiles = [.05, .10, .30, .50, .75, .95]

def fit_modelf(q):
  rslt = mood.fit(q=q)
  return [q, rslt.params['Intercept'], rslt.params['home_rank']] + \
  rslt.conf_int().loc['home_rank'].tolist()

mdels = [fit_modelf(x) for x in percentiles]
mdels = pd.DataFrame(mdels, columns=['q', 'a', 'b', 'lb', 'ub'])

ols = smf.ols('home_score ~ home_rank', fin_com).fit()
ols_ci = ols.conf_int().loc['home_rank'].tolist()
ols = dict(a = ols.params['Intercept'],
           b = ols.params['home_rank'],
           lb = ols_ci[0],
           ub = ols_ci[1])
print(mdels)
print('\n')
print(ols)
```

```
xx = np.arange(fin_com.home_rank.min(), fin_com.home_rank.max())
y_got = lambda a, b: a + b * xx

ffig, aax = plt.subplots(figsize=(10,7))

for i in range(mdels.shape[0]):
  yy = y_got(mdels.a[i], mdels.b[i])
  aax.plot(xx, yy, linestyle='dotted', color='black')

yy = y_got(ols['a'], ols['b']) 

aax.scatter(fin_com.home_rank, fin_com.home_score, alpha=.2, color='cyan')
aax.plot(xx, yy, color='red', label='OLS')

legend = aax.legend()
aax.set_xlabel('Home Team Rank', fontsize=15)
aax.set_ylabel('Goals Scored by Home Team', fontsize=15)
```

As can be seen from the output, the quantile regression model does not suffer from underfitting as much as the linear regression model does.



```
n = mdels.shape[0]
plt.figure(figsize=(9,7))
p1 = plt.plot(mdels.q, mdels.b, color='black', label='Quantile Reg.')
p2 = plt.plot(mdels.q, mdels.ub, linestyle='dotted', color='black')
p3 = plt.plot(mdels.q, mdels.lb, linestyle='dotted', color='black')
p4 = plt.plot(mdels.q, [ols['b']] * n, color='red', label='OLS')
p5 = plt.plot(mdels.q, [ols['lb']] * n, linestyle='dotted', color='red')
p6 = plt.plot(mdels.q, [ols['ub']] * n, linestyle='dotted', color='red')
plt.ylabel(r'$\beta_{home rank}$')
plt.xlabel('Quantiles of Goals Scored by Home Team distribution')
plt.legend(loc='best')
plt.show()
```

```
# specifying our predictor variables and dependent variable
from sklearn.preprocessing import StandardScaler

fiichaas = f_com[['tournament', 'year', 'month', 'home_rank', 'away_rank', 'home_team', 'away_team']]
leeboos = f_com['status']

# scaling the data to mitigate bias towards higher magnitude variables
scaler = StandardScaler()

# standardizing the data
scaler.fit(fiichaas)

# transforming the data
fiichaas = scaler.transform(fiichaas)
```

Lasso model shrinks the coefficients of all the independent variables used to zero.

The score of the model is 0, implying the independent variables do not perfectly explain variation in the dependent model

```
# Splitting the data into training and test sets, building Lasso, 
# and choosing the regularization parameter with the help of GridSearch. 
# For that, we have to define the set of parameters for GridSearch. 
# In this case, a model with the highest R-squared score will give us the best parameters.
# 

# Making necessary imports, split data into training and test sets, and choose a set of parameters 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")

# splitting features and target variables into training and test sets
X_train, X_test, y_train, y_test = train_test_split(fiichaas, leeboos, test_size=0.25, random_state=202)
parameters = {'alpha': np.concatenate((np.arange(0.1,2,0.1), np.arange(2, 5, 0.5), np.arange(5, 25, 1)))}

# instantiating the linear and lasso regression models and the gridsearch object.
linear = linear_model.LinearRegression()
lasso = linear_model.Lasso() 
gridlasso = GridSearchCV(lasso, parameters, scoring ='r2') 

# Fitting models and print the best parameters, R-squared scores, MSE, and coefficients
gridlasso.fit(X_train, y_train) 
linear.fit(X_train, y_train) 

print("lasso best parameters:", gridlasso.best_params_) 
print("\nlasso score:", gridlasso.score(X_test, y_test))
print("linear score:", linear.score(X_test, y_test)) 
print("\nlinear RMSE:", np.sqrt(mean_squared_error(y_test, linear.predict(X_test))))
print("Optimized Lasso RMSE:", np.sqrt(mean_squared_error(y_test, gridlasso.predict(X_test))))
print("\nlasso MSE:", mean_squared_error(y_test, gridlasso.predict(X_test)))
print("linear MSE:", mean_squared_error(y_test, linear.predict(X_test))) 
print("\nlasso best estimator coef: \n", gridlasso.best_estimator_.coef_)
print("\nlinear coef: \n", linear.coef_)
```

The best alpha parameter for the lasso model as determined by grid search is alpha = 0.1
